```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
```

Zadanie 1

```python
sig=[1,2,3,1]
mi=[0,1,-1,2]
x=np.linspace(-8,8,300)
wyn=[]
for i in range(4):
    wyn.append(scipy.stats.norm.pdf(x,mi[i],sig[i]))
    name="Normal dist, mi="+str(mi[i])+' sigma=',str(sig[i])
    plt.plot(x,wyn[i],label=name)
plt.legend(loc='upper center',bbox_to_anchor=(1.05, 1))
plt.show()
```

```python
rozmiar=1000
for i in range(4):
    probka=np.random.normal(mi[i],sig[i],rozmiar)
    plt.hist(probka,bins=25,histtype='bar',density=True)
    plt.plot(x,wyn[i],color='red')
    plt.title('mi:'+str(mi[i])+' sigma='+str(sig[i]))
    plt.show()
    
```

Zadanie 2

```python
#rozkład dwumianowy

n=[20,30,50,70]
p=[0.5,0.2,0.7,0.9]
rozmiar=1000
plt.rcParams["figure.figsize"] = (8,4)  
for i in range(4):
    k=range(n[i]+1)
    x=scipy.stats.binom.pmf(k,n[i],p[i])
    name="Binomial dist, n="+str(n[i])+' p=',str(p[i])
    plt.plot(k,x,'.',label=name)
plt.legend(loc='upper center',bbox_to_anchor=(1.4, 1))
plt.show()

plt.rcParams["figure.figsize"] = (10,15)  
fig,ax=plt.subplots(4)
for i in range(4):
    k=range(n[i]+1)
    x=scipy.stats.binom.pmf(k,n[i],p[i])
    y=np.random.binomial(n[i],p[i],rozmiar)
    ax[i].plot(k,x,'ro')
    ax[i].hist(y,bins=(10),density=True,stacked=True)
    ax[i].set_title('n='+str(n[i])+' p='+str(p[i]))   
plt.show()
```

```python
#rozkład wykładniczy
l=[0.5,1,1.5,2]
x=np.linspace(0,5,300)
plt.rcParams["figure.figsize"] = (8,4)  
for i in range(4):
    k=range(n[i]+1)
    wyn=scipy.stats.expon.pdf(x,scale=1/l[i])
    name="Exponential dist,  lambda="+str(l[i])
    plt.plot(x,wyn,label=name)
plt.legend(loc='upper center')
plt.show()


plt.rcParams["figure.figsize"] = (10,15) 
fig,ax=plt.subplots(4)
for i in range(4):
    wyn=scipy.stats.expon.pdf(x,scale=1/l[i])
    y=np.random.exponential(1/l[i],rozmiar)
    ax[i].plot(x,wyn,color='red')
    ax[i].hist(y,bins=(20),density=True,stacked=True)
    ax[i].set_title('lambda='+str(l[i]))
    #ax[i].title('n='+str(n[i])+' p='+str(p[i]))  
plt.show()
```

```python
# rozkład geometryczny
p=[0.2,0.3,0.5,0.7]
rozmiar=3000
plt.rcParams["figure.figsize"] = (8,4)  
for i in range(4):
    k=range(1,40)
    x=scipy.stats.geom.pmf(k,p[i])
    name="Geometrical dist, p="+str(p[i])
    plt.plot(k,x,'.',label=name)
plt.legend(loc='upper center')
plt.show()

plt.rcParams["figure.figsize"] = (10,15) 
fig,ax=plt.subplots(4)
for i in range(4):
    k=range(1,40)
    x=scipy.stats.geom.pmf(k,p[i])
    y=np.random.geometric(p[i],rozmiar)
    ax[i].plot(k,x,'ro')
    ax[i].hist(y,bins=(25//(i+1)),density=True,stacked=True)
    ax[i].set_title(' p='+str(p[i]))
plt.show()
```

```python
#rozkład Poissna
l=[10,15,20,25]
rozmiar=3000
plt.rcParams["figure.figsize"] = (8,4)  
for i in range(4):
    k=range(1,40)
    x=scipy.stats.poisson.pmf(k,l[i])
    name="Poisson dist, lambda="+str(l[i])
    plt.plot(k,x,'.',label=name)
plt.legend(loc='upper center')
plt.show()


plt.rcParams["figure.figsize"] = (10,15)  
fig,ax=plt.subplots(4)
for i in range(4):
    k=range(1,41)
    x=scipy.stats.poisson.pmf(k,l[i])
    y=np.random.poisson(l[i],rozmiar)
    ax[i].plot(k,x,'ro')
    ax[i].hist(y,bins=(20),density=True,stacked=True)
    ax[i].set_title(' lambda='+str(l[i]))
plt.show()
```

```python
#rozkład jednostajny
a=[-1,-3,-5,-10]
b=[1,2,5,20]
x=np.linspace(-20,30,1000)
plt.rcParams["figure.figsize"] = (8,4)  
for i in range(4):
    wyn=scipy.stats.uniform.pdf(x,loc=a[i],scale=(b[i]-a[i]))
    name="Uniform dist,  a="+str(a[i])+' b='+str(b[i])
    plt.plot(x,wyn,label=name)
plt.legend(loc='upper right')
plt.show()

plt.rcParams["figure.figsize"] = (10,15)  
fig,ax=plt.subplots(4)
rozmiar=3000
for i in range(4):
    wyn=scipy.stats.uniform.pdf(x,loc=a[i],scale=(b[i]-a[i]))
    y=np.random.uniform(a[i],b[i],rozmiar)
    ax[i].plot(x,wyn,color='red')
    ax[i].hist(y,bins=(20),density=True,stacked=True)
    ax[i].set_title('a='+str(a[i])+' b='+str(b[i]))
plt.show()
```

```python
#rozkład beta
a=[0.5,3,2,5]
b=[0.5,1,5,2]
x=np.linspace(0,1,300)
plt.rcParams["figure.figsize"] = (8,4)  
for i in range(4):
    wyn=scipy.stats.beta.pdf(x,a[i],b[i])
    name="Beta dist,  a="+str(a[i])+' b='+str(b[i])
    plt.plot(x,wyn,label=name)
plt.legend(loc='upper right')
plt.show()

plt.rcParams["figure.figsize"] = (10,15) 
rozmiar=3000
fig,ax=plt.subplots(4)
x=np.linspace(0,1,300)
for i in range(4):
    wyn=scipy.stats.beta.pdf(x,a[i],b[i])
    y=np.random.beta(a[i],b[i],rozmiar)
    ax[i].plot(x,wyn,color='red')
    ax[i].hist(y,bins=(20),density=True,stacked=True)
    ax[i].set_title('alfa='+str(a[i])+' beta='+str(b[i]))
plt.show()
```

```python
#rozkład gamma
a=[1,3,2,10]
b=[1,1,2,0.5]
x=np.linspace(0,15,300)
plt.rcParams["figure.figsize"] = (8,4)  
for i in range(4):
    wyn=scipy.stats.gamma.pdf(x,a[i],scale=b[i])
    name="Beta dist,  kształt="+str(a[i])+' skala='+str(b[i])
    plt.plot(x,wyn,label=name)
plt.legend(loc='upper right')
plt.show()

plt.rcParams["figure.figsize"] = (10,15)  
rozmiar=3000
fig,ax=plt.subplots(4)
for i in range(4):
    wyn=scipy.stats.gamma.pdf(x,a[i],scale=b[i])
    y=np.random.gamma(a[i],b[i],rozmiar)
    ax[i].plot(x,wyn,color='red')
    ax[i].hist(y,bins=(20),density=True,stacked=True)
    ax[i].set_title('kształt='+str(a[i])+' skala='+str(b[i]))
plt.show()
```
